# Tratamento dos Dados

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    'A': ['a0', 'a1', 'a2', 'a3'],
    'B': ['b0', 'b1', 'b2', 'b3'],
    'C': ['c0', 'c1', 'c2', 'c3'],
    'D': ['d0', 'd1', 'd2', 'd3']
}, index=[0,1,2,3])

df1
```

```python
df2 = pd.DataFrame({
    'A': ['a4', 'a5', 'a6', 'a7'],
    'B': ['b4', 'b5', 'b6', 'b7'],
    'C': ['c4', 'c5', 'c6', 'c7'],
    'D': ['d4', 'd5', 'd6', 'd7']
}, index=[4,5,6,7])

df2
```

```python
df3 = pd.DataFrame({
    'A': ['a8', 'a9', 'a10', 'a11'],
    'B': ['b8', 'b9', 'b10', 'b11'],
    'C': ['c8', 'c9', 'c10', 'c11'],
    'D': ['d8', 'd9', 'd10', 'd11']
}, index=[8,9,10,11])

df3
```

```python
pd.concat([df1, df2, df3]) # default axis=0 - VERTICALMENTE
```

```python
df4 = pd.DataFrame({
    'E': ['e0', 'e1', 'e2', 'e3'],
    'F': ['f0', 'f1', 'f2', 'f3'],
    'G': ['g0', 'g1', 'g2', 'g3'],
    'H': ['h0', 'h1', 'h2', 'h3']
}, index=[0,1,2,3])

df4
```

```python
pd.concat([df1, df4], axis=1) # Concatenação Horizontal
```

```python
esq = pd.DataFrame({
    'key1': ['k0', 'k0', 'k1', 'k2'],
    'key2': ['k0', 'k1', 'k0', 'k1'],
    'A': ['a0', 'a1', 'a2', 'a3'],
    'B': ['b0', 'b1', 'b2', 'b3']
})

esq
```

```python
dir = pd.DataFrame({
    'key1': ['k0', 'k1', 'k1', 'k2'],
    'key2': ['k0', 'k0', 'k0', 'k0'],
    'C': ['c0', 'c1', 'c2', 'c3'],
    'D': ['d0', 'd1', 'd2', 'd3']
})

dir
```

```python
pd.merge(esq, dir, on=['key1', 'key2'])
```

```python
pd.merge(esq, dir, how='outer', on=['key1', 'key2'])
```

```python
pd.merge(esq, dir, how='left', on=['key1', 'key2'])
```

```python
pd.merge(esq, dir, how='right', on=['key1', 'key2'])
```

```python
dfe = pd.DataFrame(
    {
        'A': ['a0', 'a1', 'a2'],
        'B': ['b0', 'b1', 'b2']
    },
    index=['K0', 'K1', 'K2']
)

dfe
```

```python
dfd = pd.DataFrame(
    {
        'C': ['c0', 'c1', 'c2'],
        'D': ['d0', 'd1', 'd2']
    },
    index=['K0', 'K2', 'K3']
)

dfd
```

```python
dfe.join(dfd)
```

```python
dfe.join(dfd, how='right')
```

```python
dfe.join(dfd, how='outer')
```
